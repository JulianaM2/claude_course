```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-opus-4-8"
```

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }

    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }

    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )

    return message.content[0].text
```

```python
# Interactive chat loop with user input

def execute_chat(messages):
    print("")
    question = input()

    add_user_message(messages, question)
    
    if(question.lower() == "exit" or question == ""):
        return False
    else:
        print("> " + question)
        chat_response = chat(messages)
        print("----")
        print(chat_response)
        print("----")

        add_assistant_message(messages, chat_response)
        return True


messages = []
user_wants_to_continue = True

while user_wants_to_continue:
    user_wants_to_continue = execute_chat(messages)
```
