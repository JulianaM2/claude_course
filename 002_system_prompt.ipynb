```python
# Load env variables and create client
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()
model = "claude-opus-4-8"
```

```python
# Helper functions

def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }

    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }

    messages.append(assistant_message)

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if(system is not None):
        params["system"] = system

    message = client.messages.create(**params)

    return message.content[0].text
```

```python
messages = []

system="""
You are a python engineer and you will provide a Python function.
Be as specific and concise as possible. Do not provide any additional information.
"""

add_user_message(messages, "Write a Python function that checkls a string for duplicate characters")
chat_answer = chat(messages, system=system)

chat_answer
```
